```python
import requests
from bs4 import BeautifulSoup
```

```python
url =  "https://www.flipkart.com/search?q=headphones&sid=0pm%2Cfcn%2Cgc3%2Cka8&as=on&as-show=on&otracker=AS_QueryStore_OrganicAutoSuggest_1_4_na_na_na&otracker1=AS_QueryStore_OrganicAutoSuggest_1_4_na_na_na&as-pos=1&as-type=RECENT&suggestionId=headphones%7CWireless+Headphones&requestId=d9536375-d1d7-4cd7-9835-f32d675ef56f&as-searchtext=head"
headers = {
    "User-Agent": (
        "Mozilla/5.0 (Windows NT 10.0; Win64; x64) "
        "AppleWebKit/537.36 (KHTML, like Gecko) "
        "Chrome/115.0.0.0 Safari/537.36"
    ),
    "Referer": "https://www.flipkart.com"
}
response = requests.get(url, headers=headers)
response
```

```python
if response.status_code==200:
    print('we can scrap the data')
else:
    print('we cannot scrap the data')
```

```python
soup = BeautifulSoup(response.content, "html.parser")
soup
```

```python
name=[]
price=[]
rating=[]
```

```python
product=soup.find_all("a",class_="pIpigb")

for i in product:
    name.append(i.get_text(strip=True))
```

```python
product=soup.find_all("div",class_="hZ3P6w")

for i in product:
    price.append(i.get_text(strip=True))
```

```python
product=soup.find_all("div",class_="a7saXW WDsrYC")

for i in product:
    rating.append(i.get_text(strip=True))
```

```python
import pandas as pd

# Make lists the same length
max_len = max(len(name), len(price), len(rating))

# Pad shorter lists with None or empty string
name   += [""] * (max_len - len(name))
price  += [""] * (max_len - len(price))
rating += [""] * (max_len - len(rating))

# Create DataFrame
df = pd.DataFrame({
    "Product Name": name,
    "Price": price,
    "Rating": rating
})

# Save to CSV
df.to_csv("flipkart_products.csv", index=False)

print("CSV saved successfully!")
```
